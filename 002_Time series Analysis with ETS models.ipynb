# Time series Analysis with ETS models

```r
library(fpp2)
library(readxl) 
library(tidyverse)
library(forecast)
```

```r
#read data to R and plot 
myts <- read_xlsx("RetailData.xlsx", skip=3) %>% pull("28397371") %>% 
  ts(start = c(1982,4), frequency = 12)
```

### Q1 Describing its components select an ETS model you think is appropriate for forecasting.

```r
#plot time series 
autoplot(myts)+ 
  ggtitle("Timeseries plot:
          Monthly Turnover of other recreational goods retailing in Victoria State, Australia")+
  xlab("year")+
  ylab("$million") 
```

ets(M,Ad,M) model is appropriate for forecasting this time series for the below reasons,
1. Additive trend with damped: Series increasing and then decreasing over the time and decreasing trend over the last few years was flattening out.
2. Multiplicative seasonality: Series has a strong seasonal pattern and the seasonal variation increasing as the level of the series increases.
3. Multiplicative error: There is multiplicative seasonality exists in the time series that a relative error component which is multiplicative error will be appropriated for the model.

### Q2 Fit the ETS model described in Question 1 plot the residuals and comment on these and Perform some diagnostic checks and comment on whether you are satisfied with the fit of the model. 

```r
#fit the ets model  (MAM with damped trend)
fit1 <- myts %>% ets(model = "MAM", damped = T) 
#check residuals and model's parameters, AICc, prediction accuracy
fit1 %>% autoplot()
checkresiduals(fit1)
summary(fit1)
residuals(fit1)
```

* These graphs show that the ets(M,Ad,M) produces forecasts that not appear to account for all available information. The mean of the residuals is very close to zero. However, the time plot of the residuals shows that the variation of the residuals is not much consistent across the historical data. And the ACF plot shows that there are correlations between residuals.
* The significant spikes at lags 3~7 indicate some trend information left in the residuals and at lag 12 indicates some seasonality left in the residuals that both not been well captured by the model. Histogram of the residuals shows that Bell-shape curve around zero, but long tail for the distribution implies the residuals not follow a standard normal distribution. Ljung-Box test shows that the p-value is much smaller than 0.05 which rejects the null hypothesis for the white noise.
* In sum, the model has not adequately captured the information in the time series data. Hence, I’m not satisfied with the fit of the model.

### Q3 Let R select an ets model.

```r
#fit the ets model 
fit2 <- myts %>% ets 
#check residuals and model's parameters, AICc, prediction accuracy
checkresiduals(fit2)
summary(fit2)
```

* ets(M,A,M) has been chosen by R. The ets(M,A,M) model has Multiplicative error, Additive trend and Multiplicative seasonality.
* The ets function estimates the model parameters and returns information about the fitted model. By default, it uses the AICc to select an appropriate model. The AICc information criteria including the likelihood of the model and a penalty on the total number of parameters and observations which is used for model selection. The function compares combinations of (Error, Trend, Seasonal) to choose the combination with the smallest ATCc as the best model for the forecasting. ets function chose ets(M,A,M) as it has the smallest AICc (3596.081), compared to other models.

### Q4 Comment on how the model chosen by R is different to the model you have specified and do model selection with also checking the residuals.

```r
checkresiduals(fit1)
summary(fit1)
```

```r
checkresiduals(fit2)
summary(fit2)
```

* The difference between the two models is that the ets function selected a non-damped model called ets(M,A,M) while I chose the model with damped trend called ets(M,Ad,M) .
* Both models have information left in the residuals and the properties of the residuals look similar too. However, I chose ets(M,A,M) as a better fit model because of shorter long tails for ets(M,A,M) shown in the histogram of the residuals which implies the residuals for ets(M,A,M) are more close to standard normal distribution. And ets(M,A,M) has smaller information criteria in AIC, AICc and BIC than ets(M,Ad,M).

### Q5 Generate forecasts for the period 2016-2017 using your chosen ETS model. 

```r
#use fitted model for the 2 years forecasting and plot, ets(M,A,M)
fore2 <- forecast(fit2, h = 12*2)
autoplot(fore2)+
  ggtitle("Forecast from ets(M,A,M): 
          Monthly Turnover of other recreational goods retailing in Victoria State, Australia")+
  xlab("year")+
  ylab("$million")
```

* ets(M,A,M) returns a little increasing trend for the two-year forecasting.
* When I was answering Question 1, I observed a decreasing trend over the last few years and expected a continuously decreasing in 2016-2017 with a more and more flat line. However, when I examined the series with the prediction data together in the plot, I found that it is more plausible to have a small increasing trend in the short term future instead of decreasing trend. The last 2-year data has shown an increasing trend (see December over last 2 years) which was ignored by me at the first time as it is hard to observe by just seeing the plot. The ets(M,A,M), however, did well on capturing this little increasing trend and has returned it to its forecasting.

#### Q6 Forecasts and forecast intervals generated for the period 2016-2017 from both models you have considered so far. Do both sets of forecasts look plausible? 

```r
#use fitted model for the 2 years forecasting, ets(M,Ad,M)
fore1 <- forecast(fit1, h = 12*2)
autoplot(fore1)
```

```r
#select time seris of last 10 years 
myts2006_2015 <- subset(myts, start = length(myts)-12*10+1)

#plot the selected time series with forecasting value of 2016-2017 from both ets model
autoplot(myts2006_2015)+
  autolayer(fore1, series = "forecast from ets(M,Ad,M)", alpha = 1)+
  autolayer(fore2, series = "forecast from ets(M,A,M)", alpha = 0.4)+
  ggtitle("Monthly Turnover of other recreational goods retailing in Victoria State, Australia")+
  xlab("year")+
  ylab("$million")
```

* ets(M,A,M) returns a small increasing trend for the forecast, while ets(M,Ad,M), on the other hand, returns a more stable trend with a very little decreasing trend. 
* ets(M,A,M), hence, looks more plausible as it captures the very little increasing trend exhibited over the last 3 years. In addition, ets(M,Ad,M) ‘s prediction interval wider than ets(M,A,M), indicating a more uncertainty for the prediction. These shows that ets(M,A,M) fit better for this time series than ets(M,Ad,M) does.
* However, from the diagnosis of residuals, I concluded that both models do not work well for the given time series. They were able to get part of information from data but still left information not been captured for the trend and seasonality components.
